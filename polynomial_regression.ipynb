```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.DataFrame({
    'Temperature':[1,2,3,4,5,6,7,8],
    'Energy Consumption':[2,5,9,15,14,10,5,1]
})
sns.scatterplot(data = df, x= 'Temperature', y = "Energy Consumption")
```

```python
x_poly = PolynomialFeatures(degree = 2).fit_transform( df[[("Temperature")]])
x_poly
```

```python
x = df[["Temperature"]]
y = df[["Energy Consumption"]]
x_poly = PolynomialFeatures(degree = 2).fit_transform(x)
model = LinearRegression().fit(x_poly,y)
print(f"coeff:{model.coef_}")
print(f"intercept:{model.intercept_}")

```

```python
energy_consumption_prediction = model.predict(x_poly)
print(f"energy_consumption_prediction:{energy_consumption_prediction}")
```

```python
sns.scatterplot(data = df,x = "Temperature", y = "Energy Consumption")
plt.plot (x, energy_consumption_prediction )
```
