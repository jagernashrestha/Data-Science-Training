```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    'X':[1400,1600,1700,1875,1100,1550,2350,2450,1425,1700],
    'Y':[245,312,279,308,199,219,405,324,319,225]
})
df
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.scatterplot(data = df, x ='X', y = 'Y')
```

```python
from sklearn.linear_model import LinearRegression
print(df[['X']])
model = LinearRegression().fit(df[['X']],df[['Y']])
print(f"intercept(c):{model.intercept_[0]:2f}")
print(f"cofeficient(m):{model.coef_[0][0]:2f})")
```

```python
predicted_house_price = model.predict([[1800]])
print(predicted_house_price)
```

```python
import matplotlib.pyplot as plt
y_all_predicted = model.predict(df[['X']])
print(y_all_predicted)
sns.scatterplot(data = df,x = 'X', y = 'Y')
sns.lineplot (x = df['X'], y = y_all_predicted.flatten(), color = 'red')
# plt.plot(df['X'],y_all_predicted,color = "green")
```
