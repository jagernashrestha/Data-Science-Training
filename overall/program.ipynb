```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
corr = iris[['sepal_length','sepal_width']].corr()
corr
```

```python
print(iris[['petal_length','petal_width']].corr())
```

```python
iris.corr(numeric_only= True)
```

```python
sns.scatterplot(data = iris, x = "petal_length", y = "petal_width", hue = "species")
plt.title("petal_length vs petal_width")
plt.savefig("iris_scatterplot.png") #it saves the plot in image format
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.heatmap(corr, annot = True, cmap = "coolwarm")
plt.title("Iris Correlation Heatmap")
plt.savefig("heatmap.png")
plt.show()
```
